```python
import collections
import logging
import os
import pathlib
import re
import string
import sys
import time
import json

import numpy as np
import matplotlib.pyplot as plt

import tensorflow_text as text
import tensorflow as tf

from itertools import islice
```

```python
logging.getLogger('tensorflow').setLevel(logging.ERROR)  # suppress warnings
```

```python
def take(n, iterable):
    "Return first n items of the iterable as a list"
    return list(islice(iterable, n))
```

```python
with open('/content/drive/MyDrive/dataset/GermanQuAD/GermanQuAD/GermanQuAD_train.json') as f:
  data = json.load(f)

data = data['data']

def get_data():
  for i in data:
    yield i['paragraphs'][0]['qas'][0]['question'], i['paragraphs'][0]['qas'][0]['answers'][0]['text']
```

```python
ds = tf.data.Dataset.from_generator(get_data, output_types=(tf.string, tf.string))
```

```python
for q, a in ds.batch(3).take(1):
  print(q.numpy())

  print()
  
  print(a.numpy())
```
